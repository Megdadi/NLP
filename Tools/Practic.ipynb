```python
import numpy as np
import pandas as pd

import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
details = {
    'Name' : ['Ankit', 'Aishwarya', 'tr', 'rr', 'Shaurya', 'Shivangi'],
    'Age' : [23, 21, 22, 21,6, 33],
    'University' : ['BHU','he' , 'JNU oo', 'DU','the', 'BHU oo'],
}
  
# creating a Dataframe object 
df = pd.DataFrame(details)
print(df)
```

```python
University = df.University.str.cat(sep=' ') # concatnate all rows wethin that in particular column and seperat words by ' '
print(University)
```

```python
#function to split text into word
tokens = word_tokenize(University)
print(tokens)
print(len(tokens))
frequency = nltk.FreqDist(tokens)
print(frequency)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
tokens = [w for w in tokens if not w in stop_words]
print(tokens)
```
