# Matcher:  NLP ربط الكلمات المختلفة ذات المدلول و المعنى الواحد حتى يدرك  
# اننا نقصد نفس المعنى

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

```python
doc = nlp(u'The Solar Power industry continues to grow as demand for solarpower increases. Solar-power cars are gaining popularity.')

pattern_1 = [{'LOWER': 'solarpower'}] # denotes to solarpower
pattern_2 = [{'LOWER': 'solar'}, {'LOWER': 'power'}] # denotes  to Solar Power
pattern_3 = [{'LOWER': 'solar'}, {'IS_PUNCT': True}, {'LOWER': 'power'}]# denotes  to Solar-power

matcher.add('SolarPower', None, pattern_1, pattern_2, pattern_3)
```

```python
found_matches = matcher(doc)
print(found_matches)
print('\n #####################\n #####################')
for a,b,c in  found_matches : 
    print(f'Word ID {a} , starts at {b} & ends at {c} , and word is {doc[b:c]}')
```

```python
doc = nlp('''
يتميز الذكاءالإصطناعي أنه يسير بسرعة كبيرة نحو المستقبل

و قد بدأ الكثير من الطلاب في دراسة الذكاء الإصطناعي

و تزداد فرص العمل في الذكاء الصناعي في كل مكان


''')

A1 = [{'LOWER': 'الذكاءالإصطناعي'}]
A2 = [{'LOWER': 'الذكاء'}, {'LOWER': 'الإصطناعي'}]
A3 = [{'LOWER': 'الذكاء'},  {'LOWER': 'الصناعي'}]

matcher.add('الذكاء الإصطناعي', None, A1, A2, A3)
found_matches = matcher(doc)
print(found_matches)
```
