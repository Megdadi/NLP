# **Lemmatization refers to reducing the word to its root form as found in a dictionary, and considers the meaning of the word**



play.png

```python
import nltk
from textblob import Word
nltk.download('wordnet')
word1 = Word("apples")
print("apples:", word1.lemmatize())

word2 = Word("media")
print("media:", word2.lemmatize())

word3 = Word("greater")
print("greater:", word3.lemmatize("a"))
```

```python
text="corpora  rocks better apples media greater trouble troubling troubled friendship friendships railroads destabilize"

import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
nltk.download('punkt')
token= word_tokenize(text)
print(token)
tok=WordNetLemmatizer()
for i in token:
    print(tok.lemmatize(i))
    
  

```

# **Stemming just removes the last few characters of a word, often leading to incorrect meanings and spelling**

```python
from nltk.stem import PorterStemmer
text="corpora  rocks better apples media greater trouble troubling troubled friendship friendships railroads destabilize"

token= word_tokenize(text)
porter= PorterStemmer()
for i in token:
  print(porter.stem(i))
```
